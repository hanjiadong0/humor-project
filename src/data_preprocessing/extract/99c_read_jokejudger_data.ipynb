```python
import pandas as pd
import json

def json_to_table(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
        
        # Convert JSON data to a DataFrame
        df = pd.DataFrame(data)
        return df
    except Exception as e:
        print(f"Error: {e}")

file_path = "expunations_annotated_full.json"  
expunations = json_to_table(file_path)
print (expunations)
```

## Read the original jokes from semeval with respective codes

```python
import pandas as pd
import xml.etree.ElementTree as ET
import glob

def xml_to_dataframe(file_paths):
    data = []
    
    for file_path in file_paths:
        try:
            tree = ET.parse(file_path)
            root = tree.getroot()
            
            for text in root.findall("text"):
                text_id = text.get("id")
                sentence = " ".join(word.text for word in text.findall("word"))
                data.append((text_id, sentence))
        
        except Exception as e:
            print(f"Error processing {file_path}: {e}")
    
    df = pd.DataFrame(data, columns=["Code", "Sentence"])
    print(df)
    return df

# Example usage
file_paths = glob.glob("semeval/*.xml")  # Replace with the list of XML file paths
semeval = xml_to_dataframe(file_paths).drop_duplicates()
```

```python
semeval['Code']
```

## Join the two dataframes over the column 'code'

```python

def merge_dataframes(df1, df2, left_on, right_on):
    return df1.merge(df2, left_on=left_on, right_on=right_on, how='inner')

df_merged = merge_dataframes(expunations, semeval, "ID", "Code")
```

```python
df_merged
```

## Add column Funniness Average

```python
df_merged['Funniness Average'] = df_merged['Funniness (1-5)'].apply(lambda x: sum(x) / len(x))
```

```python
df_merged
```

```python
for i in range(10):
    print("===", df_merged['Funniness Average'][i])
    print(df_merged['Sentence'][i])
    for k in range(len(df_merged['Natural language explanation'][i])):
        print("      - ", df_merged['Natural language explanation'][i][k])
    print()
```

## Add column "Filtered Funniness", with a list of intengers: the values from the "Funniness (1-5)" column where the corresponding values in "Understand the text?" were not zero. In other words, we keep only the rating marked by who said that understood the text.


```python
df_merged['Filtered Funniness'] = df_merged.apply(
    lambda row: [y for x, y in zip(row['Understand the text?'], row['Funniness (1-5)']) if x == 1], 
    axis=1
)
```

```python
df_merged
```

## Calculate the Average of the filtered funniness column (i.e. an average rating that takes into account only the ones who said they understood the text)

```python
df_merged['Filtered Funniness Average'] = df_merged['Filtered Funniness'].apply(
    #lambda x: sum(x) / len(x))
    lambda x: np.mean(x) if len(x) > 0 else None  # You can replace None with np.nan if needed
)
```
